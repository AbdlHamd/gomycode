# Votre Objectif

#Instructions
 

#Utilisez la fonction 'input' pour inviter le joueur à entrer sa réponse sous forme de chaîne de caractères.
#Utilisez la fonction 'int' pour convertir la réponse du joueur en un nombre entier.
#Utilisez la fonction "random.randint()" pour générer un nombre entier aléatoire entre 1 et 100.
#Utilisez une instruction "if/else" pour vérifier si l'estimation du joueur est trop élevée, trop basse ou correcte. 

```python
import random

print ("Bienvenue dans le jeu Devinez le nombre!\nJe pense à un nombre entre 1 et 100. Pouvez-vous le deviner ?")

x = 0

y = random.randint(1,100)

while x != y :
  x = int(input("Entrez votre réponse : "))
  if x == y:
    print ("Félicitations ! Vous avez deviné le nombre correctement!")
  elif x < y:
    print ("Votre réponse est trop basse. Réessayez.")
  elif x > y:
    print ("Votre réponse est trop élevée. Réessayez.")
```
